```python
import spacy
from spacy.lang.en.stop_words import STOP_WORDS
from string import punctuation
```

```python
text=" India is a country located in South Asia, with a population of over 1.3 billion people. It is the second-most populous country in the world, and the seventh-largest by land area. The capital of India is New Delhi, and the official languages are Hindi and English.India has a diverse and rich history, with a mix of different cultures and traditions. It is the birthplace of several major religions, including Hinduism, Buddhism, Jainism, and Sikhism. The country has a long and complex cultural heritage, with a variety of art, music, literature, and cuisine that reflects its diverse regional and historical influences.India is a democratic republic, with a federal parliamentary system of government. The country is divided into 28 states and 9 union territories, each with its own elected government. The economy of India is the third-largest in the world in terms of purchasing power parity, and it is a major player in the global economy. The country is home to a large number of multinational corporations and has a rapidly growing technology sector.India is a diverse and fascinating country, with a rich history and a bright future. It is a land of many cultures, languages, and traditions, and it is a place where people from all walks of life can come together and thrive."
```

```python
## stop words

sw=list(STOP_WORDS)
sw
```

```python
nlp=spacy.load("en_core_web_sm")
```

```python
##load the text
doc=nlp(text)
```

```python
#tokenization

tokens=[token.text for token in doc]
print(tokens)
```

```python
#remove pinctuation

punctuation = punctuation+"\n" # add new line in the list of punctuation
punctuation
```

```python
#word frequency

wf={}
for word in doc:
    if word.text.lower() not in sw: #not stop word
        if word.text.lower() not in punctuation: #not punctuation
            if word.text not in wf.keys(): # not present in word frequency earlier
                wf[word.text]=1
            else:
                wf[word.text]+=1
```

```python
print(wf)
```

```python
max_fre=max(wf.values())
max_fre
```

```python
#normalize

for word in wf.keys():
    wf[word]=wf[word]/max_fre
```

```python
wf
```

```python
max_fre=max(wf.values())
max_fre
```

```python
# sentance tokenix=zation


```

```python
st=[sent for sent in doc.sents]
```

```python
st
```

```python
# sentace score

ss={}

for sent in st:
    for word in sent:
        if word.text.lower() in wf.keys():
            if sent not in ss.keys():
                ss[sent]=wf[word.text.lower()]
            else:
                ss[sent]=wf[word.text.lower()]    
```

```python
ss
```

```python
# 30% score

from heapq import nlargest
```

```python
# select the length

sl=int(len(st)*0.3) # 0.3 for 30%
sl
```

```python
summary=nlargest(sl,ss,key=ss.get)
```

```python
summary
```

```python
#combine the sentances

final_summary=[word.text for word in summary]
```

```python
final_summary# this is a list
```

```python
summary=" ".join(final_summary)
```
